```python
# 주요 이스케이프 문자
print("첫 번째 줄\n두 번째 줄")  # \n: 줄바꿈
print("탭\t간격\t조정")  # \t: 탭
print("큰따옴표 출력: \"Hello\"")  # \": 큰따옴표
print("작은따옴표 출력: \'Hi\'")  # \': 작은따옴표
print("백슬래시 출력: \\")  # \\: 백슬래시
```
