```python
product_name, product_price, product_amount = input("상품명, 단가, 수량을 공백기준으로 입력해주세요 : ").split()

product_amount_price = int(product_price) * int(product_amount)

product_price_amount = product_price + ", " + product_amount

print(f"{"상품명":<10}{"가격, 수량":^8}{"금액":>10}")

print(f"{product_name:<10}{product_price_amount:^8}{product_amount_price:>15}")
```
