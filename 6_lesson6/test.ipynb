```python
#quiz. bmi 구현
#자신의 몸무게(kg)를 키의(m)제곱으로 나눈값

height = float(input("자신의 키를 입력해주세요(cm) : "))
m_height = height/100
weight = float(input("자신의 몸무게를 입력해주세요(kg) : "))
bmi = round(weight/(m_height*m_height),2)

print(f"당신의 bmi는 {bmi} 입니다.")

if 30 <= bmi : 
    print(f"2단계 비만입니다.")
elif 25 <= bmi:
    print(f"1단계 비만입니다.")
elif 23 <= bmi:
    print(f"위험체중 입니다.")
elif 18.5 <= bmi:
    print(f"정상 입니다.")
elif bmi < 18.5:
    print(f"저체중 입니다.")
```
