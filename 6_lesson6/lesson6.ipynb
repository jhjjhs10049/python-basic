```python
# 최종 금액이 20000원 이상이면 배송비 무료 
product_count = int(input("주문수량 : "))
product_one_price = int(input("개당단가 : "))
product_total_price = product_count * product_one_price
delivery_price = 3000

if product_total_price >= 20000 : 
    print(f"2만원 이상 결제금액으로 배송비가 제외된 최종결제 금액은 {product_total_price}원 입니다.")
else :
    print(f"최종결제 금액은 {product_total_price}원에 배송비{delivery_price}원이 추가된 {product_total_price + delivery_price}원 입니다.")
    print(f"{20000-product_total_price}원을 추가하면 배송비가 0원이 됩니다!")
```

```python
from datetime import datetime
now = datetime.now()

# 카페 모닝 할인 시스템
print("☕ === 카페 주문 시스템 === ☕\n")

# 주문 정보
drink = input("음료: ")
price = int(input("가격: "))
hour = int(now.hour)
#hour = int(input("현재 시간 (0-23): "))

print("\n" + "=" * 40)
print("주문 내역")
print("=" * 40)
print(f"음료: {drink}")
print(f"정가: {price:,}원")

# 모닝 할인 (오전 6시 ~ 10시)
if hour >= 6 and hour < 10:
    discount = int(price * 0.3)  # 30% 할인
    price = price - discount
    print(f"모닝 할인 30%: -{discount:,}원")

print(f"최종 가격: {price:,}원")
print("=" * 40)

if hour >= 6 and hour < 10:
    print("🌅 좋은 아침입니다!")   
```

```python
# 배달비 계산 시스템
print("🚚 === 배달 주문 시스템 === 🚚\n")

# 주문 정보
restaurant = input("식당 이름: ")
menu = input("메뉴: ")
price = int(input("주문 금액: "))

# 배달비 계산
delivery_fee = 3000

print("\n" + "=" * 40)
print("주문 내역")
print("=" * 40)
print(f"식당: {restaurant}")
print(f"메뉴: {menu}")
print(f"주문 금액: {price:,}원")

# 무료 배달 조건 확인 (3만원 이상)
if price >= 30000:
    delivery_fee = 0
    print(f"배달비: 무료 (30,000원 이상 주문)")
else:
    print(f"배달비: {delivery_fee:,}원")
    print(f"※ 30,000원 이상 주문시 배달비 무료")

total = price + delivery_fee
print("-" * 40)
print(f"총 결제 금액: {total:,}원")
print("=" * 40)
```

```python
#수를 입력받아 짝수, 홀수 판별 양수, 음수 판별
num = int(input("수를 입력해주세요 : "))
print(f"입력하신수는 {num}")

if num%2 == 0 :
    print(f"짝수이자",end=", ")
else :
    print(f"홀수이자",end=", ")

if 0 < num :
    print(f"양수입니다.")
elif 0 == num :
    print(f"0입니다.")
else :
    print(f"음수입니다.")
```

```python
# 지하철 요금
# 19살 이하 청소년 750, 성인 1500, 70세이상 무료
while True :
    age = int(input("본인의 나이를 입력해주세요 : "))
    subway_fee = 0
    if 0 <= age <= 19 :
        subway_fee = 750
        print(f"고객님은 청소년이시며, 요금은 {subway_fee}원 입니다.")
        break
    elif 19 < age < 70 :
        subway_fee = 1500
        print(f"고객님은 성인이시며, 요금은 {subway_fee}원 입니다.")
        break
    elif 70 <= age :
        print(f"고객님은 70세 이상이시므로, 요금은 {subway_fee}원 입니다.")
        print(f"출퇴근 시간에는 탑승을 자제해 주십시오.")
        break
    else : 
        print(f"나이를 다시 확인해 주십시오.")
```

```python
# 날씨별 옷차림 추천 시스템
print("🌡️ === 옷차림 추천 서비스 === 🌡️\n")

# 날씨 정보 입력
location = input("지역: ")
temperature = int(input("현재 기온(°C): "))
is_raining = input("비 오나요? (yes/no): ")

print("\n" + "=" * 40)
print(f"📍 {location} 날씨 정보")
print("=" * 40)
print(f"기온: {temperature}°C")

# 온도에 따른 옷차림 추천
if temperature >= 28:
    clothes = "반팔, 반바지, 원피스"
    print(f"\n🔥 매우 더워요! ({temperature}°C)")
elif temperature >= 23:
    clothes = "얇은 긴팔, 면바지, 블라우스"
    print(f"\n☀️ 따뜻해요! ({temperature}°C)")
elif temperature >= 17:
    clothes = "니트, 가디건, 청바지"
    print(f"\n🍂 선선해요! ({temperature}°C)")
elif temperature >= 10:
    clothes = "재킷, 가죽점퍼, 기본 코트"
    print(f"\n🍃 쌀쌀해요! ({temperature}°C)")
elif temperature >= 5:
    clothes = "코트, 히트텍, 기모 바지"
    print(f"\n❄️ 추워요! ({temperature}°C)")
else:
    clothes = "패딩, 두꺼운 코트, 목도리, 장갑"
    print(f"\n⛄ 매우 추워요! ({temperature}°C)")

print(f"추천 옷차림: {clothes}")

# 우산 추천
if is_raining == "yes":
    print("\n☔ 우산을 꼭 챙기세요!")

print("=" * 40)
```
