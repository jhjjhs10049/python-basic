```python
# 유튜버 등급 판정 시스템
print("📺 === 유튜버 통계 시스템 === 📺\n")

# 채널 정보
channel_name = input("채널 이름: ")
subscribers = int(input("구독자 수: "))
views = int(input("총 조회수: "))
videos = int(input("영상 개수: "))

# 평균 조회수
avg_views = views / videos if videos > 0 else 0

# 등급 판정
if subscribers >= 1000000:
    tier = "💎 다이아몬드"
    message = "축하합니다! 최상위 유튜버입니다!"
elif subscribers >= 100000:
    tier = "🏆 골드"
    message = "훌륭해요! 인기 유튜버입니다!"
elif subscribers >= 10000:
    tier = "🥈 실버"
    message = "잘하고 있어요! 성장 중입니다!"
elif subscribers >= 1000:
    tier = "🥉 브론즈"
    message = "좋은 시작이에요!"
else:
    tier = "🌱 새싹"
    message = "화이팅! 꾸준히 업로드하세요!"

# 결과 출력
print("\n" + "╔" + "═" * 48 + "╗")
print("║" + f"{channel_name} 채널 통계".center(48) + "║")
print("╠" + "═" + "═" * 48 + "╣")
print(f"║  구독자: {subscribers:,}명".ljust(50) + "║")
print(f"║  총 조회수: {views:,}회".ljust(50) + "║")
print(f"║  영상 개수: {videos}개".ljust(50) + "║")
print(f"║  평균 조회수: {avg_views:,.0f}회".ljust(50) + "║")
print("╠" + "─" * 50 + "╣")
print(f"║  등급: {tier}".ljust(50) + "║")
print(f"║  {message}".ljust(50) + "║")

# 다음 등급까지
if subscribers < 1000:
    next_goal = 1000 - subscribers
    print(f"║  브론즈까지 {next_goal:,}명 남았어요!".ljust(50) + "║")
elif subscribers < 10000:
    next_goal = 10000 - subscribers
    print(f"║  실버까지 {next_goal:,}명 남았어요!".ljust(50) + "║")
elif subscribers < 100000:
    next_goal = 100000 - subscribers
    print(f"║  골드까지 {next_goal:,}명 남았어요!".ljust(50) + "║")
elif subscribers < 1000000:
    next_goal = 1000000 - subscribers
    print(f"║  다이아몬드까지 {next_goal:,}명 남았어요!".ljust(50) + "║")

print("╚" + "═" * 50 + "╝")
```

```python
# 영화 예매 시스템
print("🎬 === 영화 예매 시스템 === 🎬\n")

# 영화 정보
movie = input("영화 제목: ")
rating = input("관람등급 (전체/12세/15세/청불): ")

# 고객 정보
name = input("\n예매자 이름: ")
age = int(input("나이: "))
is_student = input("학생인가요? (yes/no): ")

# 예매 정보
time = int(input("\n상영 시간 (0-23시): "))
seat_type = input("좌석 등급 (일반/VIP): ")
tickets = int(input("티켓 수: "))

# 기본 가격
base_price = 14000

# 좌석 등급별 가격
if seat_type == "VIP":
    base_price = 18000

# 나이별 할인
if age < 13:
    base_price = 10000  # 어린이
elif age >= 65:
    base_price = 11000  # 경로우대
elif is_student == "yes":
    base_price = base_price - 2000  # 학생 할인

# 조조 할인 (6시-10시)
morning_discount = 0
if time >= 6 and time < 10:
    morning_discount = 2000
    base_price = base_price - morning_discount

# 총 가격
total_price = base_price * tickets

# 관람 가능 여부 확인
can_watch = True
reason = ""

if rating == "청불" and age < 18:
    can_watch = False
    reason = "청소년 관람불가 영화입니다."
elif rating == "15세" and age < 15:
    can_watch = False
    reason = "15세 미만은 관람할 수 없습니다."
elif rating == "12세" and age < 12:
    can_watch = False
    reason = "12세 미만은 관람할 수 없습니다."

# 결과 출력
print("\n" + "=" * 50)
if can_watch:
    print("✅ 예매 완료!")
    print("=" * 50)
    print(f"영화: {movie} ({rating})")
    print(f"예매자: {name} ({age}세)")
    print(f"상영시간: {time}:00")
    print(f"좌석: {seat_type} × {tickets}석")
    print("-" * 50)
    print(f"티켓 가격: {base_price:,}원 × {tickets}매")
  
    if morning_discount > 0:
        print(f"조조 할인: -{morning_discount:,}원")
  
    if is_student == "yes" and age >= 13 and age < 65:
        print(f"학생 할인 적용")
  
    print("-" * 50)
    print(f"총 결제 금액: {total_price:,}원")
    print("=" * 50)
    print("\n🍿 즐거운 관람 되세요!")
else:
    print("❌ 예매 불가!")
    print("=" * 50)
    print(f"사유: {reason}")
    print("=" * 50)
```

```python
# 조건:
# - 오전(6-11시): 아메리카노 추천
# - 오후(12-17시): 카페라떼 추천  
# - 저녁(18-23시): 디카페인 추천
# - 더운 날씨(25도 이상): 아이스 추천
# - 추운 날씨(15도 미만): 핫 추천

# hour = int(input("몇시인지 입력하십시오 (ex:1~24) : "))
from datetime import datetime
now = datetime.now()

hour = now.hour
tem = int(input("몇도인지 입력하십시오 : "))

if (6 <= hour and hour <= 11) :
    print(f"오전이니 아메리카노를 추천드립니다.")
elif (12 <= hour and hour <= 17) :
    print(f"오후이니 카페라떼를 추천드립니다.")
elif (18 <= hour and hour <= 23) :
    print(f"저녁이니 디카페인을 추천드립니다.")
else :
    print(f"영업시간이 아닙니다.")

if (25 <= tem) :
    print(f"날이 더우니 아이스는 어떠신가요")
elif (tem < 15) :
    print(f"날이 추우니 핫은 어떠신가요")
```
