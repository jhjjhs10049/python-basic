```python
# 힌트
text = input("단어를 입력하세요: ")

# 소문자로 변환하고 공백 제거
cleaned = text.lower().replace(' ', '')

# 회문 검사 (역순과 비교)
is_palindrome = cleaned == cleaned[::-1]

print(f"입력한 단어: {text}")
print(f"정리된 단어: {cleaned}")
print(f"역순: {cleaned[::-1]}")
print(f"회문 여부: {is_palindrome}")
```

```python
# 문제 26 (🔴 고급).
# 사용자로부터 전화번호를 입력받아(예: 01012345678 또는 010-1234-5678) 세 가지 형식으로 변환하여 출력하세요.
# 요구사항:
# 사용자가 하이픈(-)을 넣든 안 넣든 똑같이 처리되어야 합니다. (힌트: 먼저 replace("-", "")로 모든 하이픈을 지우고 시작하세요!)

# 형식 1: 010-1234-5678 (하이픈 재삽입)
# 형식 2: (010) 1234-5678
# 형식 3: +82-10-1234-5678 (앞의 '0' 하나를 떼고 '+82'를 붙입니다)

tel_user = input("전화번호를 입력해 주세요(예: 01012345678 또는 010-1234-5678) : ")
tel_user = tel_user.replace('-', '')
tel_user1 = tel_user[0:3]
tel_user2 = tel_user[3:7]
tel_user3 = tel_user[7:]
print(f"{tel_user1}-{tel_user2}-{tel_user3}")
print(f"({tel_user1}) {tel_user2}-{tel_user3}")
print(f"{tel_user1.replace('0', '+82-', 1)}-{tel_user2}-{tel_user3}")
```
